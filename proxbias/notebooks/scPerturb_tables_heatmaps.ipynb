Let's display Table 1 in the paper. For Table 2, you need to include the CRISPRi datasets ("ReplogleWeissman2022_rpe1", "TianKampmann2021_CRISPRi", "AdamsonWeissman2016_GSM2406681_10X010") in the `fnames` variable above and run `generate_specific_loss_and_summary_tables` for these.
```python
allres = pd.read_csv(sc.get_specific_loss_file_path())
display(allres)
```

Let's display Table S2 in the paper. Note that since we only included the two CRISPR-cas9 datasets in the `fnames`, the table displayed here will only include these.
